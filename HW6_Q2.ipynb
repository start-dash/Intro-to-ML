Part B

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
```

```python
# Load CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
```

```python
# Preprocess data
train_images = train_images / 255.0
test_images = test_images / 255.0
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
# Building Model + 2 Hidden Layers
expanded_model = Sequential([
    Flatten(input_shape=(32, 32, 3)),
    Dense(512, activation='relu'),
    Dense(256, activation='relu'),  # First additional hidden layer
    Dense(128, activation='relu'),  # Second additional hidden layer
    Dense(10, activation='softmax')
])
```

```python
# Compiling Extended Model
expanded_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Training Extended Model for 300 Epochs
import time
start_time_extended = time.time()
history_extended = expanded_model.fit(train_images, train_labels, epochs=300, batch_size=64, validation_data=(test_images, test_labels))
end_time_extended = time.time()
training_time_extended = end_time_extended - start_time_extended
```

```python
# Evaluate the extended model
test_loss_extended, test_accuracy_extended = expanded_model.evaluate(test_images, test_labels)
```

```python
# Report training time, loss, and evaluation accuracy after 300 epochs
# Analyze the results in your report
print(f"Training time: {training_time_extended} seconds")
print(f"Training loss: {history_extended.history['loss'][-1]}")
print(f"Evaluation accuracy: {test_accuracy_extended}")
```
